```python
import pandas as pd
```

```python
neurons = [
        'L5_uniform_13e9',
        'L5_shemesh_supfig9b_exp_yoff_13e9',
        'L5_shemesh_supfig9b_exp_lin_yoff_13e9',
        'L5_uniform_13e10',
        'L5_shemesh_supfig9b_exp_yoff_13e10',
        'L5_shemesh_supfig9b_exp_lin_yoff_13e10',
        'L23_uniform_13e9',
        'L23_shemesh_supfig9b_exp_yoff_13e9',
        'L23_shemesh_supfig9b_exp_lin_yoff_13e9',
        'L23_uniform_13e10',
        'L23_shemesh_supfig9b_exp_yoff_13e10',
        'L23_shemesh_supfig9b_exp_lin_yoff_13e10'
]
hoc_files = [
    'L5',
    'L5',
    'L5',
    'L5',
    'L5',
    'L5',
    'L23',
    'L23',
    'L23',
    'L23',
    'L23',
    'L23',
]
ChR_distribution = [
    'uniform',
    'shemesh_supfig9b_exp_yoff',
    'shemesh_supfig9b_exp_lin_yoff',
    'uniform',
    'shemesh_supfig9b_exp_yoff',
    'shemesh_supfig9b_exp_lin_yoff',
    'uniform',
    'shemesh_supfig9b_exp_yoff',
    'shemesh_supfig9b_exp_lin_yoff',
    'uniform',
    'shemesh_supfig9b_exp_yoff',
    'shemesh_supfig9b_exp_lin_yoff',
]
ChR_soma_density = [ # in channels / cm2
    13e9,
    13e9,
    13e9,
    13e10,
    13e10,
    13e10,
    13e9,
    13e9,
    13e9,
    13e10,
    13e10,
    13e10,
]
df = pd.DataFrame(
    columns=['neuron'],
    data=neurons
)
df['hoc'] = hoc_files
df.loc[(df.hoc=='L5'),'cortical_depth'] = 1170 # um
df.loc[(df.hoc=='L23'),'cortical_depth'] = 400 # um
df['ChR_soma_density'] = ChR_soma_density
df['ChR_distribution'] = ChR_distribution
df.set_index('neuron').to_hdf('neuron_data.hdf', key='first')
```
