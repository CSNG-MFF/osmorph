```python
# format
# dimensions (A4:  8.27 x 11.69 inches)
text_frac = 0.8
abs_pw = 8.27 # pagewidth
abs_pl = 11.69 # pagelength
pw = text_frac * abs_pw
pl = text_frac * abs_pl
# others
dpi=600
```

```python
L5_data = pd.read_pickle('../../../data/2022-08-04_L5/all_data.pickle')
L23_data = pd.read_pickle('../../../data/2022-08-04_L23/all_data.pickle')
# select relevant data from data frame
ChR_expr_lvl = 10354945 # this is 130um-2 at soma for uniform distrib
chanrhod_distribution = 'uniform'
# light model and params
light_model = 'foutz_et_al2012'
diam = 200
NA = 0.22 
# other
stim_duration = 200 # ms
# apply selection
L5_data = L5_data.loc['L5',light_model,chanrhod_distribution,
                      ChR_expr_lvl,diam, NA, stim_duration]
L23_data = L23_data.loc['L23',light_model,chanrhod_distribution,
                      ChR_expr_lvl,diam, NA, stim_duration]

light_powers = [1e-4, 3e-4, 1e-3, 5e-3]
```

```python
for lp in light_powers:
    dfplot = L5_data.loc[lp].groupby(['radius [um]']).mean().reset_index()
    plt.plot(dfplot['radius [um]'], dfplot['firint_rate [Hz]'], color='#550000ff', lw=2)
    dfplot = L23_data.loc[lp].groupby(['radius [um]']).mean().reset_index()
    plt.plot(dfplot['radius [um]'], dfplot['firint_rate [Hz]'], color='#ff00ffff', lw=2)
    plt.xlim(0,600)
    plt.ylim(0,140)
    plt.show()
```

```python
def p_thresh(thresh, distance, data, delta_dist):
    # round to distance idx
    distance_idx = delta_dist * (distance//delta_dist + (distance%delta_dist>0.5*delta_dist))
    values_per_angle = data.loc[distance_idx]['firint_rate [Hz]'].values
    return sum((values_per_angle>=thresh) / len(values_per_angle))

def area_within_ring(rmin_um,rmax_um):
    return np.pi * (rmax_um**2-rmin_um**2)

def n_neurons_within_ring(rmin_um, rmax_um, surface_density_PERum2, thresh_Hz, data):
    if rmin_um > 0:
        dist_um = (rmin_um+rmax_um)/2
        delta_dist_um = rmax_um - rmin_um
    elif rmin_um == 0:
        dist_um = 0
        delta_dist_um = rmax_um*2
    else:
        raise ValueError
    # active areas in ring (delta_dist) centered around dist_um
    area = area_within_ring(rmin_um, rmax_um) * p_thresh(thresh_Hz, dist_um, data, delta_dist_um)
    return area * surface_density_PERum2
    
def n_neurons_within_distance(max_dist_um, delta_dist_um, surface_density_PERum2, data, thresh_Hz):
    # initial circle:
    n_neurons = n_neurons_within_ring(rmin_um=0, rmax_um=delta_dist_um/2, 
                                      surface_density_PERum2=surface_density_PERum2, 
                                      thresh_Hz=thresh_Hz, data=data)
    # following rings:
    for dist in np.arange(delta_dist_um, max_dist_um+delta_dist_um,delta_dist_um):
        n_neurons += n_neurons_within_ring(
            rmin_um=dist-delta_dist_um, 
            rmax_um=dist+delta_dist_um, 
            surface_density_PERum2=surface_density_PERum2, 
            thresh_Hz=thresh_Hz, data=data
        )
    return n_neurons

n_neurons_within_distance(max_dist_um=200, delta_dist_um=25, surface_density_PERum2=1, data=L5_data.loc[lp], thresh_Hz=100)
```

```python
thresholds = np.arange(5,155,5)
distances = np.arange(0,625,25)
for lp in light_powers:
    for thresh in [5, 50, 100]:
        L5_n_activated = [
            p_thresh(thresh=thresh, distance=distance, data=L5_data.loc[lp], delta_dist=25)\
            for distance in distances
        ]
        plt.plot(distances, L5_n_activated, label=str(thresh))    
    plt.ylim(0,None)
    plt.xlim(0,None)
    plt.legend(title='threshold response [Hz]')
    plt.show()
```

```python
# plot activation along cortical surface
rng = np.random.default_rng()

def rnd_smpled_fr_at_distance(data, distance, delta_dist, density_reduction_factor):
    # round to distance idx
    distance_idx = delta_dist * (distance//delta_dist + (distance%delta_dist>0.5*delta_dist))
    fr_over_angles = data.loc[distance_idx]['firint_rate [Hz]'].values
    # spare out firing rate to account for reduced neural density:
    results = [0] * int(density_reduction_factor)
    results[0] = rng.choice(fr_over_angles) 
    return rng.choice(results)

def cortical_activation_map(data, delta_dist, width_um, length_um, neuron_density_PERmm2, density_reduction_factor):
    
    neurons_along_length = int(np.sqrt(neuron_density_PERmm2 * length_um * 1e-3 / (width_um * 1e-3)))
    neurons_along_width = int(np.sqrt(neuron_density_PERmm2 * width_um * 1e-3 / (length_um * 1e-3)))
    l_neuron = length_um / neurons_along_length
    w_neuron = length_um / neurons_along_width
    fr_neurons = []
    for i in range(neurons_along_length):
        fr_neurons.append(
            [rnd_smpled_fr_at_distance(data, distance=np.sqrt((l_neuron*i-length_um/2)**2+(w_neuron*j-width_um/2)**2), delta_dist=delta_dist, density_reduction_factor=density_reduction_factor) for j in range(neurons_along_width)]
        )
    return fr_neurons
```
