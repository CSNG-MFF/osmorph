```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_hdf('../../../data/conductance_prof/all_data.hdf')
df = df.reset_index()[['light_power', 'radius [um]','angle [rad]','AP_count','total_int_g','total_peak_g','max_int_sec_g']]
df
```

```python
df['norm_total_int_g'] = df.groupby(['light_power'])['total_int_g'].apply(lambda x: x/x.max())
df['norm_total_peak_g'] = df.groupby(['light_power'])['total_peak_g'].apply(lambda x: x/x.max())
df['norm_max_int_sec_g'] = df.groupby(['light_power'])['max_int_sec_g'].apply(lambda x: x/x.max())
df = df.set_index(['light_power', 'radius [um]', 'angle [rad]'])
```

```python
df
```

```python
pd.DataFrame(df.loc[5e-4].norm_total_int_g - df.loc[5e-5].norm_total_int_g).reset_index()
```

```python
df = df.loc[:,:300,:]
```

```python
import matplotlib.colors
from neurostim.polarmaps import simple_polar_map
import matplotlib as mpl
import matplotlib.pyplot as plt
import copy
lps = df.reset_index().light_power.unique()
lps = [5e-5,5e-4,2e-2]

fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
ax, mappable = simple_polar_map(
        data=pd.DataFrame(df.loc[5e-4].norm_total_int_g - df.loc[5e-5].norm_total_int_g).reset_index(),
        plot_col='norm_total_int_g',
        cmap=copy.copy(mpl.cm.get_cmap("coolwarm")),
        norm = matplotlib.colors.DivergingNorm(
            vmin=-0.15,
            vcenter=0,
            vmax=0.15),
        ax=ax,
        vmin=None,
        vmax=None)
plt.colorbar(mappable)
plt.title('5e-4 - 5e-5')
plt.show()
# fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
# ax, mappable = simple_polar_map(
#         data=pd.DataFrame(df.loc[5e-4].total_int_g - df.loc[5e-5].total_int_g).reset_index(),
#         plot_col='total_int_g',
#         cmap=copy.copy(mpl.cm.get_cmap("coolwarm")),
#         ax=ax,
#         vmin=None,
#         vmax=None)
# plt.colorbar(mappable)
# plt.title(str(lp))
# plt.show()

for lp in lps:
    fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
    ax, mappable = simple_polar_map(
            data=df.loc[lp].reset_index(),
            plot_col='AP_count',
            cmap=copy.copy(mpl.cm.get_cmap("Reds")),
            ax=ax,
            vmin=0,
            vmax=25)
    #plt.colorbar(mappable)
    plt.title(str(lp))
    plt.savefig('plots/tmp_fr_'+str(lp)+'.png', dpi=300)
    plt.show()
# fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
# ax, mappable = simple_polar_map(
#         data=pd.DataFrame(df.loc[2e-2].norm_total_int_g - df.loc[5e-4].norm_total_int_g).reset_index(),
#         plot_col='norm_total_int_g',
#         cmap=copy.copy(mpl.cm.get_cmap("coolwarm")),
#         norm = matplotlib.colors.DivergingNorm(
#                 vmin=-0.5,
#                 vcenter=0,
#                 vmax=0.5),
#         ax=ax,
#         vmin=None,
#         vmax=None)
# plt.colorbar(mappable)
# plt.show()
# for lp in lps:
#     fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
#     ax, mappable = simple_polar_map(
#             data=df.loc[lp].reset_index(),
#             plot_col='norm_total_int_g',
#             cmap=copy.copy(mpl.cm.get_cmap("jet")),
#             ax=ax,
#             vmin=0,
#             vmax=1)
#     #plt.colorbar(mappable)
#     plt.title(str(lp))
#     plt.savefig('plots/tmp_cond_'+str(lp)+'.png', dpi=300)
#     plt.show()
# for lp in lps:
#     fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
#     ax, mappable = simple_polar_map(
#             data=df.loc[lp].reset_index(),
#             plot_col='norm_total_peak_g',
#             cmap=copy.copy(mpl.cm.get_cmap("jet")),
#             ax=ax,
#             vmin=0,
#             vmax=1)
#     #plt.colorbar(mappable)
#     plt.title(str(lp))
#     plt.savefig('plots/tmp_peak_cond_'+str(lp)+'.png', dpi=300)
#     plt.show()

for lp in lps:
    fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
    ax, mappable = simple_polar_map(
            data=df.loc[lp].reset_index(),
            plot_col='norm_max_int_sec_g',
            cmap=copy.copy(mpl.cm.get_cmap("jet")),
            ax=ax,
            vmin=0,
            vmax=1)
    #plt.colorbar(mappable)
    plt.title(str(lp))
    plt.savefig('plots/tmp_max_cond_'+str(lp)+'.png', dpi=300)
    plt.show()
```
